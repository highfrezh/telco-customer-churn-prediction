Rule of Thumb for Correlation Thresholds:
- Correlation < 0.7: Generally acceptable.
- Correlation 0.7–0.85: Review and decide based on domain knowledge.
- Correlation > 0.85: Consider dropping, combining, or transforming features.

2. Check Multicollinearity for Categorical features

Null Hypothesis ($ H_0 $): The Feature is independent of target column (No-Correlation)

Alternative Hypothesis ($ H_1 $): The Feature and Target column are not independent (Correalted)

```python
from scipy.stats import chi2_contingency
chi2_test = []
for feature in categorical_features:
    if chi2_contingency(pd.crosstab(df['Churn Label'], df[feature]))[1] < 0.05:
        chi2_test.append('Reject Null Hypothesis')
    else:
        chi2_test.append('Fail to Reject Null Hypothesis')
result = pd.DataFrame(data=[categorical_features, chi2_test]).T
result.columns = ['Column', 'Hypothesis Result']
result
```

In the context of the Chi-Square test, the **null hypothesis** and its interpretations are as follows:

#### **Null Hypothesis (H₀):**
> There is no significant relationship between the categorical feature and the target variable (`Churn Label`).

#### **Alternative Hypothesis (H₁):**
> There is a significant relationship between the categorical feature and the target variable (`Churn Label`).

#### **Interpretation of Results:**

#### **1. Reject Null Hypothesis:**
- **Meaning**: There is enough evidence to conclude that the feature is **significantly related** to the target variable (`Churn Label`).
- **Implication**: The feature is important and likely contributes to explaining or predicting the target. You might want to retain this feature for your classification task.

#### **2. Fail to Reject Null Hypothesis:**
- **Meaning**: There is not enough evidence to conclude that the feature is significantly related to the target variable (`Churn Label`).
- **Implication**: The feature may not be important for your classification model. You might consider dropping it, but this depends on additional factors (e.g., domain knowledge, interaction effects with other variables).

#### **Why "Fail to Reject" instead of "Accept"?**
- Hypothesis testing does not "prove" the null hypothesis is true. Instead, it determines whether there is insufficient evidence to reject it. 
- In simpler terms, **"Fail to Reject"** means that the data does not provide strong evidence against the null hypothesis, but it doesn’t confirm the null hypothesis as absolutely true.

---

#### **What to Do Based on the Results?**

1. **Features with "Reject Null Hypothesis":**
   - Keep these features, as they have a significant relationship with the target.

2. **Features with "Fail to Reject Null Hypothesis":**
   - You might consider removing them if they do not add value to your model. However, before removing them:
     - Check if they have interactions with other features.
     - Consider their importance in feature selection techniques (e.g., recursive feature elimination, feature importance scores).
